# Dimensionality Reduction of Wine Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
# Load the Wine dataset
wine = load_wine()
X = wine.data
y = wine.target
feature_names = wine.feature_names
```

```python
X
```

```python
X.shape
```

```python
y.shape
```

```python
feature_names
```

```python
# Initialize LDA
lda = LDA(n_components=2)
lda
```

```python
# Fit LDA on the dataset
X_lda = lda.fit_transform(X, y)
X_lda
```

```python
# Create a DataFrame with LDA results
df_lda = pd.DataFrame(X_lda, columns=['LDA Component 1', 'LDA Component 2'])
df_lda['Target'] = y
```

```python
df_lda.head()
```

```python
# plotting the LDA clusters
plt.scatter(df_lda['LDA Component 1'], df_lda['LDA Component 2'])
```

```python
# Plotting
lw = 2
target_names = wine.target_names

plt.figure()
for color, i, target_name in zip(['r', 'g', 'b'], [0, 1, 2], target_names):
    plt.scatter(df_lda[df_lda['Target'] == i]['LDA Component 1'],
                df_lda[df_lda['Target'] == i]['LDA Component 2'],
                color=color, alpha=.8, lw=lw,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('LDA of Wine dataset')
plt.xlabel('LDA Component 1')
plt.ylabel('LDA Component 2')
plt.show()
```

All three clusters are completely distinct.

```python
from sklearn.decomposition import PCA
```

```python
# Initialize LDA
pca = PCA(n_components=2)
pca
```

```python
# Fit LDA on the dataset
X_pca = pca.fit_transform(X, y)
X_pca
```

```python
# Create a DataFrame with LDA results
df_pca = pd.DataFrame(X_pca, columns=['PCA Component 1', 'PCA Component 2'])
df_pca['Target'] = y
```

```python
# Plotting
lw = 2
target_names = wine.target_names

plt.figure()
for color, i, target_name in zip(['r', 'g', 'b'], [0, 1, 2], target_names):
    plt.scatter(df_pca[df_pca['Target'] == i]['PCA Component 1'],
                df_pca[df_pca['Target'] == i]['PCA Component 2'],
                color=color, alpha=.8, lw=lw,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of Wine dataset')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```

This does not look like a good clustering step. Lot's of confusion between classes.

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
scaler = StandardScaler()
```

```python
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
pca = PCA(n_components=2)
pca.fit(X_scaled)
X_pca = pca.transform(X_scaled)
```

```python
# Create a DataFrame with LDA results
df_pca = pd.DataFrame(X_pca, columns=['PCA Component 1', 'PCA Component 2'])
df_pca['Target'] = y
```

```python
# Plotting
lw = 2
target_names = wine.target_names

plt.figure()
for color, i, target_name in zip(['r', 'g', 'b'], [0, 1, 2], target_names):
    plt.scatter(df_pca[df_pca['Target'] == i]['PCA Component 1'],
                df_pca[df_pca['Target'] == i]['PCA Component 2'],
                color=color, alpha=.8, lw=lw,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA Scaled of Wine dataset')
plt.xlabel('PCA Scaled Component 1')
plt.ylabel('PCA Scaled Component 2')
plt.show()
```

This looks way better, but there is still some confusion in some of the classes.

```python
from sklearn.manifold import TSNE
```

```python
# Initialize LDA
tsne = TSNE(n_components=2)
tsne
```

```python
tsne.fit(X)
X_tsne = tsne.fit_transform(X)
```

```python
# Create a DataFrame with LDA results
df_tsne = pd.DataFrame(X_tsne, columns=['TSNE Component 1', 'TSNE Component 2'])
df_tsne['Target'] = y
```

```python
# Plotting
lw = 2
target_names = wine.target_names

plt.figure()
for color, i, target_name in zip(['r', 'g', 'b'], [0, 1, 2], target_names):
    plt.scatter(df_tsne[df_tsne['Target'] == i]['TSNE Component 1'],
                df_tsne[df_tsne['Target'] == i]['TSNE Component 2'],
                color=color, alpha=.8, lw=lw,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('TSNE of Wine dataset')
plt.xlabel('TSNE Component 1')
plt.ylabel('TSNE Component 2')
plt.show()
```

A lot of confusion between the classes. Let's scale the data.

```python
sc = StandardScaler()
scaler = StandardScaler()
```

```python
# Scale the data with z-score
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
# Define TSNE and transform
tsne = TSNE(n_components=2)
tsne.fit(X_scaled)
X_tsne = tsne.fit_transform(X_scaled)
```

```python
# Create a DataFrame with LDA results
df_tsne = pd.DataFrame(X_tsne, columns=['TSNE Component 1', 'TSNE Component 2'])
df_tsne['Target'] = y
```

```python
# Plotting
lw = 2
target_names = wine.target_names

plt.figure()
for color, i, target_name in zip(['r', 'g', 'b'], [0, 1, 2], target_names):
    plt.scatter(df_tsne[df_tsne['Target'] == i]['TSNE Component 1'],
                df_tsne[df_tsne['Target'] == i]['TSNE Component 2'],
                color=color, alpha=.8, lw=lw,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('TSNE Scaled of Wine dataset')
plt.xlabel('TSNE Scaled Component 1')
plt.ylabel('TSNE Scaled Component 2')
plt.show()
```

This clustering looks much better, but not as good as the LDA one. LDA is the best method in this case.
