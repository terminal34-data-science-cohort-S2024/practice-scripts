# Code Example of Correlation Coefficients for Feature Selection

For this purpose, we will use the “Penguins” dataset available through the Seaborn library, which is a great alternative to the Iris dataset and offers a mix of continuous and categorical variables.

Task

Explore the relationships between the flipper length, body mass, and bill length of penguins using Pearson, Spearman, and Kendall correlations. Visualize these relationships to enhance our understanding.

Step 1: Install and Import Necessary Libraries

Ensure seaborn, pandas, scipy, and matplotlib are installed:

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr, kendalltau
```

## Load penguin dataset

```python
penguins_df = sns.load_dataset('penguins').dropna()
penguins_df.head()
```

## Compute Correlations

```python
pearson_corr_flipper_body = pearsonr(
    penguins_df['flipper_length_mm'], penguins_df['body_mass_g'])[0]
print(f'Pearson Correlation between Flipper Lenghth and Body Mass {pearson_corr_flipper_body}')
```

```python
spearman_corr_flipper_body = spearmanr(
  penguins_df['flipper_length_mm'], penguins_df['body_mass_g'])[0]
print(f'Spearman Correlation between Flipper Lenghth and Body Mass {spearman_corr_flipper_body}')
```

```python
kendall_corr_flipper_body = kendalltau(
  penguins_df['flipper_length_mm'], penguins_df['body_mass_g'])[0]
print(f'Kendall Correlation between Flipper Lenghth and Body Mass {spearman_corr_flipper_body}')
```

```python
pearson_corr_flipper_bill = pearsonr(
    penguins_df['flipper_length_mm'], penguins_df['bill_depth_mm'])[0]
print(f'Pearson Correlation between Flipper Lenghth and Body Mass {pearson_corr_flipper_bill}')
```

## Negatively correlated example

```python
sns.scatterplot(penguins_df, x='flipper_length_mm', y='bill_depth_mm')
```

## Positively correlated example

```python
sns.scatterplot(penguins_df, x='flipper_length_mm', y='body_mass_g')
```

## Positively correlated example

```python
pearson_corr_flipper_bill = pearsonr(
    penguins_df['bill_length_mm'], penguins_df['body_mass_g'])[0]
print(f'Pearson Correlation between Flipper Lenghth and Body Mass {pearson_corr_flipper_bill}')
sns.scatterplot(penguins_df, x='bill_length_mm', y='body_mass_g')
```

## Perfectly correlated example

```python
pearson_corr_flipper_bill = pearsonr(
    penguins_df['body_mass_g'], penguins_df['body_mass_g'])[0]
print(f'Pearson Correlation between Flipper Lenghth and Body Mass {pearson_corr_flipper_bill}')
sns.scatterplot(penguins_df, x='body_mass_g', y='body_mass_g')
```
