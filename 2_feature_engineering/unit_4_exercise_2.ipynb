# Dimensionality Reduction of Iris Dataset


```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import pandas as pd
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
```

```python
# Initialize PCA model
pca = PCA(n_components=2)

# Fit PCA on the dataset
X_pca = pca.fit_transform(X)
```

```python
# Create a DataFrame with PCA results
df_pca = pd.DataFrame(X_pca, columns=['Principal Component 1', 'Principal Component 2'])
df_pca['Target'] = y

# Plot the PCA-transformed dataset
colors = ['navy', 'turquoise', 'darkorange']
lw = 2
target_names = iris.target_names

plt.figure()
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(df_pca[df_pca['Target'] == i]['Principal Component 1'],
                df_pca[df_pca['Target'] == i]['Principal Component 2'],
                color=color, alpha=.8, lw=lw,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of IRIS dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
