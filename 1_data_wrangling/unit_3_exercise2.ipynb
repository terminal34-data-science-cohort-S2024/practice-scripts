```python
import pandas as pd
```

```python
dsjobs_df = pd.read_csv('drive/MyDrive/Terminal34/Uncleaned_DS_jobs.csv')
dsjobs_df
```

```python
dsjobs_df.dtypes
```

```python
nodata_job_title = dsjobs_df['Job Title'].isnull().sum()
nodata_job_title
```

```python
dsjobs_df['Salary Estimate'].unique()
```

```python
dsjobs_df['Salary Estimate'] = dsjobs_df['Salary Estimate'].replace(
    {
        '\$': '',
        'K': '',
        ' \(Glassdoor est.\)': '',
        '\(Employer est.\)': ''
    }, regex=True
)
dsjobs_df['Salary Estimate'].unique()
```

```python
dsjobs_df['Salary Estimate'].value_counts().plot.bar()
```

```python
dsjobs_df['Rating'].unique()
```

```python
dsjobs_df['Rating'] = dsjobs_df['Rating'].apply(lambda x: 0 if x < 0 else x)
dsjobs_df['Rating'].unique()
```

```python
dsjobs_df['Company Name'].unique()
```

```python
dsjobs_df['Company Name'] = dsjobs_df['Company Name'].apply(
    lambda x: x.split('\n')[0] if '\n' in x else x)
```

```python
dsjobs_df
```

```python
dsjobs_df['Size'].unique()
```

```python
dsjobs_df['Size'] = dsjobs_df['Size'].replace(['-1', 'Unknown'], pd.NA)
dsjobs_df['Size'].unique()
```

```python
dsjobs_df['Founded'].unique()
```

```python
dsjobs_df['FoundedClean'] = dsjobs_df['Founded'].apply(
    lambda x: pd.NA if x < 0 else x)
dsjobs_df['FoundedClean'].unique()
```

```python
dsjobs_df['Salary Estimate'].unique()
```

```python
dsjobs_df[['Salary Min', 'Salary Max']] = dsjobs_df['Salary Estimate'].str.split(
    '-', expand=True).apply(pd.to_numeric, errors='ignore')
dsjobs_df
```
