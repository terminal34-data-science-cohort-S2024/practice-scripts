```python
science_df = science_df.iloc[1:].reset_index(drop=True)
science_df
```

```python
science_df = science_df.reset_index(drop=True)
science_df
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
science_df.groupby('Indicator Name').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# For each year where "Indicator Name" is "Scientific and technical journal articles",
# plot a time series with the Value column (Line Plot)
fig, ax = plt.subplots(figsize=(10, 8))

journal_df = science_df[science_df['Indicator Name'] == 'Scientific and technical journal articles']
#journal_df['Year'] = journal_df['Year'].astype(int)
journal_df['Value'] = journal_df['Value'].astype(float)
journal_df = journal_df.sort_values(by=['Year']).reset_index(drop=True)

sns.barplot(data=journal_df, x='Year', y='Value', ax=ax, alpha=0.5)
sns.lineplot(data=journal_df, x='Year', y='Value', ax=ax, color='red', marker='o')
#journal_df

#journal_df_bar = science_df[science_df['Indicator Name'] == 'Scientific and technical journal articles']
#journal_df_bar['Year'] = journal_df_bar['Year'].astype(str)

#plt.ylabel('Number of articles')
#plt.title('Number of articles as a function of time')
#plt.xlim(2003, 2016)
```
