```python
import numpy as np
import pandas as pd
```

```python
white_df = pd.read_csv('winequality-white.csv', sep=';')
white_df
```

```python
# find NaNs in the dataset
white_df[white_df.isna().any(axis=1)]
```

```python
white_df.isnull().sum()
```

```python
# convert quality to categorical
white_df['quality'].min(), white_df['quality'].max()
```

```python
white_df['quality'].unique()
```

```python
white_df['quality_categorical'] = pd.cut(
    white_df['quality'], bins=[0, 5, 7, 10], labels=['Low', 'Medium', 'High'])
white_df
```

```python
white_df['quality_categorical'].value_counts().plot.bar()
```

```python
white_df['quality'].value_counts().plot.bar()
```

```python
white_df.columns
```

```python
white_df['alcohol'].head()
```

```python
white_df['alcohol'].min(), white_df['alcohol'].max()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
white_df['alcohol_normalized'] = scaler.fit_transform(white_df['alcohol'].values.reshape(-1, 1))
white_df['alcohol'].min(), white_df['alcohol'].max()
```

```python
white_df['alcohol_normalized'].min(), white_df['alcohol_normalized'].max()
```

```python
# concatenate two dataframes
red_df = pd.read_csv('winequality-red.csv', sep=';')
red_df
```

```python
wine_df = pd.concat([white_df, red_df], axis=0).reset_index(drop=True)
wine_df
```

```python
wine_df['quality_categorical'] = pd.cut(
    wine_df['quality'], bins=[0, 5, 7, 10], labels=['Low', 'Medium', 'High'])
wine_df
```

```python
scaler = MinMaxScaler()
wine_df['alcohol_normalized'] = scaler.fit_transform(wine_df['alcohol'].values.reshape(-1, 1))
wine_df['alcohol'].min(), wine_df['alcohol'].max()
```
